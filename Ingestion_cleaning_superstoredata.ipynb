### Creating star schema for better exploration 

```python
# 1.Created Fact table AS sales_fact
create_table_query = """
CREATE TABLE IF NOT EXISTS sales_fact (
    order_id VARCHAR(50),
    order_date DATE,
    ship_date DATE,
    ship_mode VARCHAR(50),
    customer_id VARCHAR(50),
    product_id VARCHAR(50),
    sales DECIMAL(10,2),
    quantity INT,
    discount FLOAT,
    profit FLOAT
);
"""

insert_data_query = """
INSERT INTO sales_fact (order_id, order_date, ship_date, ship_mode, customer_id, product_id, sales, quantity, discount, profit)
SELECT 
    order_id, 
    order_date, 
    ship_date, 
    ship_mode, 
    customer_id, 
    product_id, 
    sales, 
    quantity, 
    discount, 
    profit 
FROM cleaned_raw_data;
"""

# 3. Execute the commands
with engine.connect() as conn:
    conn.execute(text(create_table_query))
    print("Table 'sales_fact' created or already exists.")
    
    conn.execute(text(insert_data_query))
    conn.commit()  
    print("Data successfully loaded into 'sales_fact'.")
```

```python
# 2.Created dimensional table as customer_dim
query_customer = """
CREATE TABLE IF NOT EXISTS customer_dim (
    customer_id VARCHAR(50),
    customer_name VARCHAR(100),
    segment VARCHAR(50),
    country VARCHAR(50),
    city VARCHAR(50),
    state VARCHAR(50),
    postal_code INT,
    region VARCHAR(50)
);
"""


insert_customer = """
INSERT INTO customer_dim
SELECT DISTINCT
    customer_id,
    customer_name,
    segment,
    country,
    city,
    state,
    postal_code,
    region
FROM cleaned_raw_data;

"""
```

```python
with engine.connect() as conn:
    conn.execute(text(query_customer))
    conn.execute(text(insert_customer))
    conn.commit() 
```

```python
# 3.Created dimensional table as product_dim
query_product = """
CREATE TABLE IF NOT EXISTS  product_dim (
    product_id VARCHAR(50),
    category VARCHAR(50),
    sub_category VARCHAR(50),
    product_name TEXT
);
"""

insert_product = """
INSERT INTO product_dim
SELECT DISTINCT
    product_id,
    category,
    sub_category,
    product_name
FROM cleaned_raw_data;
"""
```

```python
with engine.connect() as conn:
    conn.execute(text(query_product))
    conn.execute(text(insert_product))
    conn.commit() 
```
